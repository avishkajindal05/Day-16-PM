# EDA Assignment Notebook

Dataset: **eda_assignment_data.csv**

This notebook performs:
- Distribution analysis
- Relationship analysis
- Correlation analysis
- Category comparison
- Interactive visualizations with Plotly


```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from scipy import stats

sns.set(style="whitegrid")

df = pd.read_csv("eda_assignment_data.csv")
df.head()
```

## Part A — Distribution Plots

```

num_cols = df.select_dtypes(include=np.number).columns[:3]

for col in num_cols:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```


Example insight:
The histogram shows the distribution and skewness of the feature.  
KDE highlights whether the distribution is normal or skewed.


## Relationship Plots

```

num_cols = df.select_dtypes(include=np.number).columns

if len(num_cols) >= 2:
    sns.regplot(x=df[num_cols[0]], y=df[num_cols[1]])
    plt.title(f"{num_cols[0]} vs {num_cols[1]}")
    plt.show()

if len(num_cols) >= 3:
    sns.regplot(x=df[num_cols[0]], y=df[num_cols[2]])
    plt.title(f"{num_cols[0]} vs {num_cols[2]}")
    plt.show()
```


Example insight:
The regression line helps identify correlation trends between variables.  
Scatter spread indicates whether the relationship is strong or weak.


## Correlation Heatmap

```

corr = df.corr(numeric_only=True)

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

Insight: Strong correlations may indicate predictive relationships.

## Box Plot Across Categories

```

cat_col = df.select_dtypes(include='object').columns[0]
num_col = df.select_dtypes(include=np.number).columns[0]

sns.boxplot(x=df[cat_col], y=df[num_col])
plt.title(f"{num_col} across {cat_col}")
plt.xticks(rotation=45)
plt.show()
```

Insight: Box plots reveal distribution differences and outliers across categories.

## Part B — Interactive Plotly Charts

```

fig = px.histogram(df, x=num_cols[0])
fig.write_html("plotly_histogram.html")
fig.show()

fig2 = px.scatter(df, x=num_cols[0], y=num_cols[1], trendline="ols")
fig2.write_html("plotly_scatter.html")
fig2.show()

fig3 = px.box(df, x=cat_col, y=num_col)
fig3.write_html("plotly_box.html")
fig3.show()
```


Interactive charts allow:
- Zooming into dense regions of the data
- Hover tooltips revealing precise values


## Part C — Coding Question

```

def plot_numerical_eda(df):
    import matplotlib.pyplot as plt
    import scipy.stats as stats
    
    num_cols = df.select_dtypes(include=np.number).columns
    
    for col in num_cols:
        fig, ax = plt.subplots(1,3, figsize=(12,4))
        
        ax[0].hist(df[col])
        ax[0].set_title(f"Histogram: {col}")
        
        ax[1].boxplot(df[col])
        ax[1].set_title(f"Boxplot: {col}")
        
        stats.probplot(df[col], dist="norm", plot=ax[2])
        ax[2].set_title(f"QQ Plot: {col}")
        
        plt.tight_layout()
    
    plt.savefig("eda_numerics.png")
```
